## Jobs, Stages and Actions

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder.master("local[5]")\
        .appName("JobStageTest")\
        .getOrCreate()
spark
```

```python
employee_df = spark.read.format("csv")\
              .option("header",True)\
              .load("data_2.csv")
```

```python
print(employee_df.rdd.getNumPartitions())
employee_df = employee_df.repartition(2)
print(employee_df.rdd.getNumPartitions())
# Added "age" to the select statement
employee_df = employee_df.filter(col("salary") > 900000)\
                .select("id", "name", "salary", "age")\
                .groupby("age").count()
employee_df.collect()
```
