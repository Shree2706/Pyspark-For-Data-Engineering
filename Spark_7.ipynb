## Union and Union All

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder.master ("local[5]")\
        .appName("Union&UnionAll")\
        .getOrCreate()
spark
```

```python
data=[(10 ,'Anil',50000, 18),
(11 ,'Vikas',75000,  16),
(12 ,'Nisha',40000,  18),
(13 ,'Nidhi',60000,  17),
(14 ,'Priya',80000,  18),
(15 ,'Mohit',45000,  18),
(16 ,'Rajesh',90000, 10),
(17 ,'Raman',55000, 16),
(18 ,'Sam',65000,   17),
(18 ,'Sam',55000,   17),
(18 ,'Sam',65000,   17)]
```

```python
schema = ['id', 'name', 'salary', 'mngr_id']

manager_df = spark.createDataFrame(data=data, schema=schema)
```

```python
manager_df.show()
```

```python
manager_df.count()
```

```python
data1=[(19 ,'Sohan',50000, 18),
(20 ,'Sima',75000,  17)]

schema1 = ['id', 'name', 'salary', 'mngr_id']
manager_df1 = spark.createDataFrame(data=data1, schema=schema1)
```

```python
manager_df1.show()
```

```python
## 1. Union
manager_df.union(manager_df1).count()
```

```python
manager_df.createOrReplaceTempView("manager_df_tbl")
manager_df1.createOrReplaceTempView("manager_df1_tbl")
```

```python
spark.sql("""

select * from manager_df_tbl
union
select * from manager_df1_tbl

""").count()
```

```python
## Union All
spark.sql("""

select * from manager_df_tbl
union all
select * from manager_df1_tbl

""").count()
```
